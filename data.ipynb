```python
sns.countplot(data=df, x="Attrition")
plt.show()
```

```python
attrition_rate = (
    df["Attrition"].value_counts(normalize=True)*100
)

print(attrition_rate)
```

```python
role = pd.crosstab(
    df["JobRole"],
    df["Attrition"]
)

print(role)
```

```python
salary = df.groupby("Department")["MonthlyIncome"].mean()

print(salary)
```

```python
corr = df.select_dtypes(include=np.number).corr()

plt.figure(figsize=(14,10))

sns.heatmap(
    corr,
    cmap="BrBG",
    annot=True
)

plt.show()
```

```python
bins = [18,25,35,45,55,65]

labels = [
    "18-25",
    "26-35",
    "36-45",
    "46-55",
    "56-65"
]

df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=bins,
    labels=labels
)
```

```python
df["IncomeGroup"] = pd.qcut(
    df["MonthlyIncome"],
    q=4,
    labels=[
        "Low",
        "Medium",
        "High",
        "Very High"
    ]
)
```

```python
df.to_csv(
    "HR_Analytics_Cleaned.csv",
    index=False
)
```
